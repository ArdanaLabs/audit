```python
@ntrct0
def mkscatter(colidx: int): 
    clean._mkscatter(df.columns[colidx])
```

```python
@ntrct0
def mkkde(colidx: int): 
    clean._mkkde(df.columns[colidx])
```

```python
@ntrct1
def heatmap(colidx_x: int, colidx_y: int, colidx_color: int): 
    return clean._heatmap(df.columns[colidx_x], df.columns[colidx_y], df.columns[colidx_color])
```
